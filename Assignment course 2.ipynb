### 1. Dictionary

- **Dictionary code**: How to create a dictionary.
- **Dictionary method**: How to use a dictionary method called `values()`.

### A dictionary is a set of values where each value is associated with a key of access.  
**For example**: The "key 1" is associated with the value "1", the "key 2" is associated with the value "2" and the "key 3" is associated with the value "3".  We can see the values of the dictionary using the method " `values()` ".

```python
#1 create a dictionary 
D = {"key1":1, "key2":2,"key3":3, "key4":4}

#Show the values of the dictionary
D.values()
```
